```python
import matplotlib.pyplot as plt
import numpy as np
from wordcloud import WordCloud, ImageColorGenerator
```

```python
mfw = [('kommen', 562), ('sehen', 438), ('gehen', 394), ('Nacht', 357), ('Leben', 333), ('mehr', 323), ('sagen', 311), ('schon', 294), ('ganz', 289), ('Tag', 287), ('Zeit', 285), ('bleiben', 281), ('Traum', 277), ('mal', 276), ('Liebe', 273), ('wissen', 271), ('wer', 245), ('nie', 242), ('immer', 240), ('geben', 239)]
```

```python
mfw.reverse()
```

```python
mfw
```

```python
dict(mfw)
```

```python
wordcloud = WordCloud(max_words=20, background_color="white", colormap='GnBu').generate_from_frequencies(dict(mfw))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
worte = ()
counts = []
for tup in mfw:
    worte = worte + (tup[0],)
    counts.append(tup[1])
y_pos = np.arange(len(worte))
```

```python
colors = ['#DAF7A6', '#FFC300', '#FF5733', '#C70039', '#900C3F', '#581845']
```

```python
plt.title('most frequent words')
plt.ylabel('Worte')
plt.xlabel("Häufigkeit")
plt.barh(y_pos, counts, color='#81bbee')
plt.yticks(y_pos, worte)
plt.show()
```
